```python
#グラフの確認
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model

#教師データ
datasets_x = [[155],[165],[170]] #身長
datasets_y = [[48],[63],[76]] #体重（予測したい値）
plt.scatter(datasets_x,datasets_y, color="black")
plt.show()
```

```python
from sklearn import linear_model
#線形回帰モデルの呼び出し
reg = linear_model.LinearRegression()
#線形回帰モデルに教師データを学習させる
reg.fit(datasets_x,datasets_y)
```

```python
#グラフの確認
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model

#教師データ
datasets_x = [[155],[165],[170]] #身長
datasets_y = [[48],[63],[76]] #体重（予測したい値）
plt.scatter(datasets_x,datasets_y, color="black")

#直線描画用の身長データ、155~170までの数値を作成
line_x = np.linspace(155, 170)
#作成したモデルを使って、↑のデータの身長に対する体重を求める
line_y = reg.predict(line_x.reshape(-1,1))
plt.plot(line_x, line_y, color = 'blue')
plt.show()
#数式の回帰係数(coef_)と切片(intercept_)を表示
print(reg.coef_,reg.intercept_)
```
